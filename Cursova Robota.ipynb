## Conda
**Conda** - це вільно розповсюджувана система управління пакетами та середовищами для роботи з мовами програмування, такими як Python, R, Ruby, Lua, Scala, Java, JavaScript, C/C++, FORTRAN, Haskell, HTML, CSS, и XML. Conda призначена для управління залежностями пакетів, установки та оновлення пакетів, а також створення та управління середовищами.

## Можливості conda
- **Управління пакетами.** Conda дозволяє швидко і легко встановлювати, оновлювати, видаляти та управляти версіями пакетів. Всі пакети зберігаються у власних репозиторіях, і вони можуть бути легко доступні для встановлення.
- **Управління середовищами.** Conda дозволяє створювати і управляти ізольованими середовищами, що дозволяє уникнути конфліктів залежностей між різними проектами. Ви можете легко створювати, клонувати та видаляти середовища.
- **Багатомовна підтримка.** Conda підтримує багато мов програмування, що робить його корисним для проектів, які використовують різні мови.
- **Швидкість та ефективність.** Conda оптимізована для швидкої установки та оновлення пакетів. Вона також може автоматично вирішувати проблеми залежностей, що полегшує управління пакетами.
- **Інтеграція з іншими інструментами.** Conda може використовуватися разом із іншими інструментами управління пакетами, наприклад, pip.
- **Кросплатформеність.** Conda працює на багатьох операційних системах, включаючи Windows, macOS та Linux.

## Переваги conda перед pip
Обидва інструменти, conda та pip, призначені для управління пакетами та залежностями в програмному забезпеченні на Python, але вони мають свої власні особливості та переваги.

- **Управління залежностями та середовищами:**<br>
    *Conda*: Забезпечує ізольовані середовища, що дозволяє вам створювати окремі середовища для різних проектів та уникнути конфліктів між версіями пакетів.<br>
    *pip*: Залежності рішаться вручну, і ізоляція середовищ вимагає використання інших інструментів, таких як virtualenv чи venv.<br>
    
- **Мульти-мовна підтримка:**<br>
    *Conda*: Підтримує різні мови програмування, не тільки Python.<br>
    *pip*: Спеціалізується на Python та не забезпечує підтримку для інших мов.<br>
    
- **Розв'язання проблем залежностей:**<br>
    *Conda*: Має вбудований розв'язувач проблем залежностей, що полегшує установку пакетів та їх залежностей.<br>
    *pip*: Може викликати проблеми залежностей, особливо в складних проектах, і вирішення цих проблем може бути складним завданням.<br>
    
- **Швидкість та ефективність:**<br>
    *Conda*: Швидше вирішує проблеми залежностей, оскільки має вбудований розв'язувач проблем.<br>
    *pip*: Може вимагати багато часу на розв'язання складних проблем залежностей, особливо коли використовуються великі проекти.<br>
    
- **Завантаження зрізів пакетів:**<br>
    *Conda*: Дозволяє завантажувати передвстановлені бінарні пакети, що може суттєво зекономити час при установці.<br>
    *pip*: Зазвичай встановлює пакети з вихідних кодів, що може займати більше часу на компіляцію та установку.

### Розрахунок мат. сподівання та дисперсіїї.
Вирахуємо математичне сподівання та дисперсію. Порівняємо із результатами функцій **pandas**. Нормалізуємо дані.

```python
# pandas calculating variance and mean
def pd_mean_std():
    mean = df.mean(numeric_only=True)
    std = df.std(numeric_only=True)
    return mean, std
    
# custom calculations
def my_mean_std(col):
    col_list = df[col].to_list()
    elements = set(col_list)
    math_exp = sum(i*(col_list.count(i)/len(col_list)) for i in elements)
    variance = sqrt((sum((i-math_exp)**2 for i in col_list))/(len(col_list)-1))
    return math_exp, variance
    
print(f"Pandas mean for price: {pd_mean_std()[0]['price']} VS my mean for price {my_mean_std('price')[0]}") 
print(f"Pandas standart variance for price: {pd_mean_std()[1]['price']} VS my standart variance for price: {my_mean_std('price')[1]}")

# data standartization
# for i in ['price', 'mileage', 'lot', 'year']:
#     mean_std = pd_mean_std()
#     df[i] = df[i].apply(lambda x: (x- mean_std[0][i]) / mean_std[1][i])

for i in ['price', 'mileage', 'lot', 'year']:
    df[i] = (df[i] - df[i].min()) / (df[i].max() - df[i].min())
```

### Побудова різних регресійних моделей та візуалізація (Завдання 1, 4)
Використаємо класи *LinearRegression*, *RandomForestRegression* та *Multy Layer Perceptron* для побудови моделей та натренуємо їх.

```python
plt.scatter(x1_test, y1_test)
plt.scatter(x1_test, y2_pred_for)
plt.show()
```

```python
model3 = MLPRegressor()
model3.fit(x1_train, y1_train)
y3_pred_mlp = model3.predict(x1_test)
```

```python
plt.scatter(x1_test, y1_test)
plt.scatter(x1_test, y3_pred_mlp)
plt.show()
```

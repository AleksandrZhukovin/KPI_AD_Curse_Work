## Вступ 

***Jupyter Notebook (раніше IPython Notebooks)*** — це веб-інтерактивне обчислювальне середовище для створення Notebook документів. 

Документ **Jupyter Notebook** — це просте інтерактивне середовище програмування на основі браузера, що містить упорядкований список комірок введення/виводу, який може містити код, текст (за допомогою *Markdown*), математику, графіки та мультимедійні дані. Під інтерфейсом notebook — це документ *JSON*, який відповідає версійній схемі, він зазвичай закінчується розширенням «.ipynb». 

**IPython Notebook** - потужний інструмент розробки та подання проекту в області *Data Science*. Він дозволяє зберігати код, результати, візуалізації та деталізовані коментарі в одночас. Розробка програми в такому форматі дає змогу швидко перевіряти весь код або його окремі частини, мати під рукою візуалізацію результатів і оформлювати проект для презентації іншим людям завдяки мові розмітки тексту. 

Використання хмарних обчислень дозволяє ефективно виконувати код незалежно від потужностей локальної машини користувача, що в свою чергу дозволяє працювати не маючи необхідного заліза для запуску коду. 

**Anaconda** — це вільно та відкрито розповсюджуваний (freemium) дистрибутив різних програмних продуктів, зокрема, мов програмування Python та R. Платформа спеціалізується на «наукових обчисленнях» (scientific computing): наука про дані, застосуванні методів машинного навчання, широкомасштабна обробка даних, передбачувальна аналітика тощо. Використання платформи має на меті спрощення управління пакетами та їх розгортання. 

Завантаження дистрибутиву *Anaconda* виправдовується значною кількістю переваг які він надає. По перше - велика кількість під’єднуваних модулів. По друге - графічний інтерфейс на заміну звичному командному рядку. А також пропрацьований менеджер пакетів conda. Головною відмінність conda від стандартного pip є аналіз програмного середовища та встановлення пакетів з урахуванням поточних залежностей або попередження користувача про можливі проблеми із сумісністю.

Вирахуємо математичне сподівання та дисперсію. Порівняємо із результатами функцій **pandas**.

```python
# pandas calculating variance and mean
def pd_mean_std():
    mean = df.mean(numeric_only=True)
    std = df.std(numeric_only=True)
    return mean, std
    
# custom calculations
def my_mean_std(col):
    col_list = df[col].to_list()
    elements = set(col_list)
    math_exp = sum(i*(col_list.count(i)/len(col_list)) for i in elements)
    variance = sqrt((sum((i-math_exp)**2 for i in col_list))/(len(col_list)-1))
    return math_exp, variance
    
print(f"Pandas mean for price: {pd_mean_std()[0]['price']} VS my mean for price {my_mean_std('price')[0]}") 
print(f"Pandas standart variance for price: {pd_mean_std()[1]['price']} VS my standart variance for price: {my_mean_std('price')[1]}")

# data standartization
# for i in ['price', 'mileage', 'lot', 'year']:
#     mean_std = pd_mean_std()
#     df[i] = df[i].apply(lambda x: (x- mean_std[0][i]) / mean_std[1][i])

for i in ['price', 'mileage', 'lot', 'year']:
    df[i] = (df[i] - df[i].min()) / (df[i].max() - df[i].min())
```

# Частина 3

```python
x1 = df[['mileage']]
y1 = df['year']
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.4, random_state=40)
```

```python
model1 = LinearRegression()
model1.fit(x1_train, y1_train)
y1_pred = model1.predict(x1_test)
```

```python
plt.scatter(x1_test, y1_test)
plt.scatter(x1_test, y1_pred)
plt.show()
```

```python
model2 = RandomForestRegressor(n_estimators=100, max_depth=3)
model2.fit(x1_train, y1_train)
y2_pred = model2.predict(x1_test)
```

```python
plt.scatter(x1_test, y1_test)
plt.scatter(x1_test, y2_pred)
plt.show()
```

```python
model3 = MLPRegressor()
model3.fit(x1_train, y1_train)
y3_pred = model3.predict(x1_test)
```

```python
plt.scatter(x1_test, y1_test)
plt.scatter(x1_test, y3_pred)
plt.show()
```

```python
print('Linear Regression error:', mean_squared_error(y1_test, y1_pred))
print('Random Forest Regression error:', mean_squared_error(y1_test, y2_pred))
print('Multi Layer Perceptron error:'mean_squared_error(y1_test, y3_pred))
```

```python
x1_2d = pd.concat([x1, x1**2], axis=1)
x1_train, x1_test, y1_train, y1_test = train_test_split(x1_2d, y1, test_size=0.4, random_state=40)
```

```python
model_2d = RandomForestRegressor(n_estimators=100, max_depth=3)
model_2d.fit(x1_train, y1_train)
y_2d_pred = model_2d.predict(x1_test)
print('Random Forest Regression error:', mean_squared_error(y1_test, y_2d_pred))
print(x1.shape)
```
